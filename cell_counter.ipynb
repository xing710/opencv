对图像进行二值化,选取一个全局阈值，然后就把整幅图像分成了非黑即白的二值图像了。函数为cv2.threshold() 

这个函数有四个参数，第一个原图像，第二个进行分类的阈值，第三个是高于（低于）阈值时赋予的新值，第四个是一个方法
选择参数，常用的有： 
cv2.THRESH_BINARY（黑白二值） 
cv2.THRESH_BINARY_INV（黑白二值反转）
cv2.THRESH_TRUNC （得到的图像为多像素值） 
cv2.THRESH_TOZERO 
cv2.THRESH_TOZERO_INV 
该函数有两个返回值，第一个retVal（得到的阈值值（在后面一个方法中会用到）），第二个就是阈值化后的图像。

In this chapter,
We will learn different morphological operations like Erosion, Dilation, Opening, Closing etc.
We will see different functions like : 
cv2.erode(),
cv2.dilate(), 
cv2.morphologyEx() etc.
	// 形态学操作
The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object (Always try to keep foreground in white). So what does it do? The kernel slides through the image (as in 2D convolution). A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero).

```python
import cv2
import numpy as np

img = cv2.imread('j.png',0)
img_open=cv2.imread('opening.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)  #erosion 
dilation = cv2.dilate(img,kernel,iterations = 1) #dilation
opening = cv2.morphologyEx(img_open, cv2.MORPH_OPEN, kernel)
# plt.subplot(1,3,1)
# plt.imshow(img, cmap='gray')


titles = ['img','erosion','dilation','opening']
images = [img,erosion,dilation,opening]
for i in range(len(titles)):
    plt.subplot(1,len(titles),i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

plt.figure(1, figsize=(15, 9),dpi=100,)
# 
img = cv2.imread('cell1.png',0)
kernel = np.ones((5,5),np.uint8)

ret,thresh1 = cv2.threshold(img,110,250,cv2.THRESH_BINARY)
# gray = cv2.cvtColor(thresh1, cv2.COLOR_BGR2GRAY) 
opening = cv2.morphologyEx(thresh1, cv2.MORPH_OPEN, kernel)

titles = ['img','thresh1','opening']
images = [img,thresh1,opening]
for i in range(len(titles)):
    plt.subplot(1,len(titles),i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
